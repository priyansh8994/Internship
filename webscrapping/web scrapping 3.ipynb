```python
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait


chrome_options = Options()
chrome_options.add_argument("--headless")  


webdriver_path = (r"C:\Users\hp\Desktop\chromedriver_win32")


driver = webdriver.Chrome(executable_path=webdriver_path, options=chrome_options)


keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']


num_images_per_keyword = 10


for keyword in keywords:
    
    driver.get("https://images.google.com")

    
    search_bar = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, '//textarea[@class="gLFyf"]'))
    )

   
    search_bar.clear()
    search_bar.send_keys(keyword)
    search_bar.submit()

    
    time.sleep(2)
    scroll_count = 0
    while scroll_count < 5:  
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        scroll_count += 1

   
    image_elements = driver.find_elements(By.XPATH, "//img")

    
    count = 0
    for image_element in image_elements:
        image_url = image_element.get_attribute("src")
        if image_url.startswith("https"):
            response = requests.get(image_url)
            if response.status_code == 200:
               
                with open(f"{keyword}_{count+1}.jpg", "wb") as f:
                    f.write(response.content)
                count += 1
        if count == num_images_per_keyword:
            break

    print(f"Scraped {count} images for '{keyword}'.")


driver.quit()
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd


driver = webdriver.Chrome("chromedriver.exe")


driver.get("https://www.amazon.in/")


search_query = input("Enter the product to search: ")


search_input = driver.find_element(By.XPATH,'//input[@class="nav-input nav-progressive-attribute"]')
search_input.send_keys(search_query)

search_btn = driver.find_element(By.XPATH,'//div[@class="nav-search-submit nav-sprite"]')
search_btn.click()


data = []


for page in range(3):
    
    search_results = driver.find_elements(By.XPATH, '//div[@class="sg-col-inner"]')

   
    for result in search_results:
        try:
            brand = result.find_element(By.XPATH,'.//div[@class="a-section a-spacing-none"]' ).text.strip()
        except:
            brand = "-"
        try:
            title = result.find_element(By.XPATH, './/span[@class="a-size-base-plus a-color-base a-text-normal"]').text.strip()
        except:
            title = "-"
        try:
            price = result.find_element(By.XPATH, './/span[@class="a-price-whole"]').text.strip()
        except:
            price = "-"
        try:
            return_exchange = result.find_element(By.XPATH, './/span[@class="a-size-small a-color-link a-text-normal"]').text.strip()
        except:
            return_exchange = "-"
        try:
            expected_delivery = result.find_element(By.XPATH, './/a[@class="a-size-small a-link-normal a-text-normal"]').text.strip()
        except:
            expected_delivery = "-"
        try:
            availability = result.find_element(By.XPATH, '//div[@class="a-column a-span12 a-text-center icon-container scrollable-container celwidget a-declarative"]').text.strip()
        except:
            availability = "-"
        try:
            product_url = result.find_element(By.XPATH, '//span[@class="a-button-inner"]').get_attribute('href')
        except:
            product_url = "-"

        data.append({
            'Brand Name': brand,
            'Name of the Product': title,
            'Price': price,
            'Return/Exchange': return_exchange,
            'Expected Delivery': expected_delivery,
            'Availability': availability,
            'Product URL': product_url
        })

    
    try:
        next_page = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        next_page.click()
    except:
        break



df = pd.DataFrame(data)
df.to_csv('product_details.csv', index=False)

print("Data saved successfully.")
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd


driver = webdriver.Chrome("chromedriver.exe")


driver.get("https://www.flipkart.com/")


search_query = input("Enter the smartphone to search: ")


search_input = driver.find_element(By.XPATH, '//input[@class="_3704LK"]')
search_input.send_keys(search_query)


search_input.submit()


data = []


search_results = driver.find_elements(By.XPATH, '//div[@class="_1AtVbE"]')


for result in search_results:
    try:
        brand = result.find_element(By.XPATH, './/div[@class="_4rR01T"]').text
    except:
        brand = "-"
    try:
        name = result.find_element(By.XPATH, './/a[@class="IRpwTa"]').text
    except:
        name = "-"
    try:
        color = result.find_element(By.XPATH, './/div[@class="_1AtVbE"]/div[2]/div[1]/div[1]').text
    except:
        color = "-"
    try:
        specs = result.find_element(By.XPATH, './/ul[@class="_1xgFaf"]/li[1]').text
        ram = "-"
        rom = "-"
        primary_camera = "-"
        secondary_camera = "-"
        display_size = "-"
        battery_capacity = "-"
        
        
        spec_parts = specs.split("|")
        for part in spec_parts:
            part = part.strip()
            if part.startswith("RAM"):
                ram = part.replace("RAM", "").strip()
            elif part.startswith("ROM"):
                rom = part.replace("ROM", "").strip()
            elif part.startswith("Camera"):
                cameras = part.replace("Camera", "").strip().split(",")
                primary_camera = cameras[0].strip()
                if len(cameras) > 1:
                    secondary_camera = cameras[1].strip()
            elif part.startswith("Display"):
                display_size = part.replace("Display", "").strip()
            elif part.startswith("Battery"):
                battery_capacity = part.replace("Battery", "").strip()
    except:
        ram = "-"
        rom = "-"
        primary_camera = "-"
        secondary_camera = "-"
        display_size = "-"
        battery_capacity = "-"

    try:
        price = result.find_element(By.XPATH, './/div[@class="_30jeq3 _1_WHN1"]').text
    except:
        price = "-"
    try:
        product_url = result.find_element(By.XPATH, './/a[@class="IRpwTa"]').get_attribute('href')
    except:
        product_url = "-"

    data.append({
        'Brand Name': brand,
        'Smartphone Name': name,
        'Colour': color,
        'RAM': ram,
        'Storage(ROM)': rom,
        'Primary Camera': primary_camera,
        'Secondary Camera': secondary_camera,
        'Display Size': display_size,
        'Battery Capacity': battery_capacity,
        'Price': price,
        'Product URL': product_url
    })




driver.quit()


df = pd.DataFrame(data)


df.to_csv('smartphone_details.csv', index=False)

print("Data saved successfully.")

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By


driver = webdriver.Chrome("chromedriver.exe")


driver.get("https://www.google.com/maps")


city = input("Enter the city name: ")


search_input = driver.find_element(By.XPATH, '//input[@autofocus="autofocus"]')
search_input.send_keys(city)


search_input.submit()


driver.implicitly_wait(5)


try:
    coordinates_element = driver.find_element(By.XPATH, '//div[@data-tooltip="Copy coordinates"]')
    coordinates = coordinates_element.get_attribute('data-value')
    latitude, longitude = coordinates.split(',')
    latitude = latitude.strip()
    longitude = longitude.strip()

    print("Latitude:", latitude)
    print("Longitude:", longitude)
except:
    print("Coordinates not found for the given city.")

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By


driver = webdriver.Chrome(r"C:\Users\hp\Desktop\chromedriver_win32\chromedriver.exe")


driver.get("https://www.digit.in/top-products/best-gaming-laptops-40.html")


laptop_listings = driver.find_elements(By.XPATH, '//div[@class="top-serch"]')[:10]


data = []


for laptop in laptop_listings:
    name = laptop.find_element(By.XPATH, '/html/body/div[9]/div[1]/div[3]/div[2]/div[1]/div[1]/div[1]/span[2]').text
    try:
        price = laptop.find_element(By.XPATH, './/td[contains(text(), "Price")]/following-sibling::td').text
    except:
        price = "-"
    try:
        specs = laptop.find_elements(By.XPATH, './/table[@class="compare"]/tbody/tr')
        specifications = ""
        for spec in specs:
            key = spec.find_element(By.XPATH, './th').text
            value = spec.find_element(By.XPATH, './td').text
            specifications += f"{key}: {value}\n"
    except:
        specifications = "-"

    data.append({
        'Name': name,
        'Price': price,
        'Specifications': specifications
    })


for item in data:
    print(item)

# Close the browser
driver.quit()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


driver = webdriver.Chrome(r"C:\Users\hp\Desktop\chromedriver_win32\chromedriver.exe")


video_url = "https://www.youtube.com/watch?v=02jD2rCdLUA"
driver.get(video_url)


SCROLL_PAUSE_TIME = 2
last_height = driver.execute_script("return document.documentElement.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height


comments = driver.find_elements(By.XPATH, '//div[@class="style-scope ytd-comment-renderer"]')
upvotes = driver.find_elements(By.XPATH, '//div[@class="yt-spec-touch-feedback-shape__fill"]')
timestamps = driver.find_elements(By.XPATH, '//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')


data = []


for comment, upvote, timestamp in zip(comments, upvotes, timestamps):
    comment_text = comment.text.strip()
    upvote_count = upvote.text.strip()
    timestamp_text = timestamp.get_attribute("aria-label").strip()
    
    data.append({
        'Comment': comment_text,
        'Upvotes': upvote_count,
        'Timestamp': timestamp_text
    })

    
    if len(data) >= 500:
        break


for item in data:
    print(item)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By


driver = webdriver.Chrome(r"path/to/chromedriver")


driver.get("https://www.hostelworld.com/search?search_keywords=London&country=England&city=London&date_from=2023-06-01&date_to=2023-06-07")


hostel_listings = driver.find_elements(By.XPATH, '//div[@class="fabresult rounded row"]')


data = []


for hostel in hostel_listings:
    name = hostel.find_element(By.XPATH, './/h2/a').text
    distance = hostel.find_element(By.XPATH, './/span[contains(text(), "km from city centre")]').text
    ratings = hostel.find_element(By.XPATH, './/div[@class="score orange big"]//span[@class="rating rating-summary-container big"]//span[@class="score orange big"]').text
    total_reviews = hostel.find_element(By.XPATH, './/div[@class="reviews"]//div[contains(text(), "Total Reviews")]/span').text
    overall_reviews = hostel.find_element(By.XPATH, './/div[@class="reviews"]//div[contains(text(), "Overall")]/span').text
    privates_price = hostel.find_element(By.XPATH, './/div[contains(@class, "prices-col private-col")]/div[contains(@class, "price-col")]/span').text
    dorms_price = hostel.find_element(By.XPATH, './/div[contains(@class, "prices-col dorm-col")]/div[contains(@class, "price-col")]/span').text
    facilities = hostel.find_element(By.XPATH, './/div[@class="facilities-label"]').text
    description = hostel.find_element(By.XPATH, './/div[@class="rating-factors"]//div[@class="rating-factors-prop"]//div[@class="prop-card-text"]').text
    
    data.append({
        'Name': name,
        'Distance from City Centre': distance,
        'Ratings': ratings,
        'Total Reviews': total_reviews,
        'Overall Reviews': overall_reviews,
        'Privates from Price': privates_price,
        'Dorms from Price': dorms_price,
        'Facilities': facilities,
        'Property Description': description
    })


for item in data:
    print(item)

```
